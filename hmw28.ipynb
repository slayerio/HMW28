```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
'CustomerID': np.arange(1, 11),
'FirstName': ['John', 'Steve', 'Anna', 'Mike', 'Emily', 'Jake', 'Laura', 'Nick', 'Olivia',
'David'],
'LastName': ['Smith', 'Johnson', 'Williams', 'Jones', 'Brown', 'Davis', 'Miller',
'Wilson', 'Moore', 'Taylor'],
'Age': np.random.randint(18, 65, size=10),
'TotalPurchaseAmount': np.random.randint(100, 1000, size=10)
})
df
```

```python
df['classification'] = df['TotalPurchaseAmount'].apply(lambda x: 'High' if x > 500 else 'Low')
df['classification_by_age'] = np.vectorize(lambda x, y: 'Low_young' if x < 20 and y == 'Low'
                                           else 'Low_adult' if x >= 20 and y == 'Low'
                                          else 'High_young' if x < 20 and y == 'High'
                                          else 'High_adult')(df['Age'],df['classification'])
df
```
